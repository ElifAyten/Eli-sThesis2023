### DF2 Image Locations

```python
import pandas as pd

file_path = "/Users/elifayten/Downloads/Thesis-4/2023_July 20, 2023_02.11.csv"

# Read the CSV file
df2 = pd.read_csv(file_path)

df2
```

```python

# Find the index of the "Part1-Q1_1" column
index = df2.columns.get_loc("Part1-Q1_1")    #this is only for a mistake was made in the survey

# Slice the DataFrame to keep only columns from "Part1-Q1_1" onwards
df2 = df2.iloc[:, index:]

df2 = df2.iloc[6:] #delete 6 rows since they are empty values

df2=df2.drop(columns='Part1.2-Q6_1')

df2 = df2.reset_index(drop=True)


column_names = []
for i in range(1, 61):
    for j in range(1, 4):
        column_names.append(f"{i}.{j}")

df2=df2.rename(columns=dict(zip(df2.columns, column_names))) # change column names for simplicity 

df2=df2.fillna("0")#fill NaN values with 0

df2=df2.astype(str)#make each value a string

#show df2
df2
```

```python
### Changge the values in DF2 with mapping
```

```python
# X.1,2,3
for question in range(1,61):
    
    mapping={"1":df1.loc[(question-1),1],
             "2":df1.loc[(question-1),2],
             "3":df1.loc[(question-1),3]
            }
    
    df2.loc[:, f"{question}.1"] = df2.loc[:, f"{question}.1"].apply(lambda x: x.replace("1", mapping["1"]))
    df2.loc[:, f"{question}.1"] = df2.loc[:, f"{question}.1"].apply(lambda x: x.replace("2", mapping["2"]))
    df2.loc[:, f"{question}.1"] = df2.loc[:, f"{question}.1"].apply(lambda x: x.replace("3", mapping["3"]))
    
    df2.loc[:, f"{question}.2"] = df2.loc[:, f"{question}.2"].apply(lambda x: x.replace("1", mapping["1"]))
    df2.loc[:, f"{question}.2"] = df2.loc[:, f"{question}.2"].apply(lambda x: x.replace("2", mapping["2"]))
    df2.loc[:, f"{question}.2"] = df2.loc[:, f"{question}.2"].apply(lambda x: x.replace("3", mapping["3"]))
    
    
    
    df2.loc[:, f"{question}.3"] = df2.loc[:, f"{question}.3"].apply(lambda x: x.replace("1", mapping["1"]))
    df2.loc[:, f"{question}.3"] = df2.loc[:, f"{question}.3"].apply(lambda x: x.replace("2", mapping["2"]))
    df2.loc[:, f"{question}.3"] = df2.loc[:, f"{question}.3"].apply(lambda x: x.replace("3", mapping["3"]))
    
```

## Part 1 & 2 (survey) sub-questions a, b, c 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace df2 with your DataFrame
# For example:
# df2 = pd.read_csv('your_file.csv')

# Step 1: Extract columns ending with '.3 .2.1'
filtered_columns_1 = df2.iloc[:, :109].filter(regex=r'\.3$', axis=1)
filtered_columns_2 = df2.iloc[:, :109].filter(regex=r'\.2$', axis=1)
filtered_columns_3 = df2.iloc[:, :109].filter(regex=r'\.1$', axis=1)



# Step 2: Count the occurrences of each model for each set of questions
count_dall_e_1 = filtered_columns_1.apply(lambda x: x.str.contains('DALL-E').sum())
count_deepdream_1 = filtered_columns_1.apply(lambda x: x.str.contains('DeepDream').sum())
count_dreamstudio_1 = filtered_columns_1.apply(lambda x: x.str.contains('DreamStudio').sum())

count_dall_e_2 = filtered_columns_2.apply(lambda x: x.str.contains('DALL-E').sum())
count_deepdream_2 = filtered_columns_2.apply(lambda x: x.str.contains('DeepDream').sum())
count_dreamstudio_2 = filtered_columns_2.apply(lambda x: x.str.contains('DreamStudio').sum())

count_dall_e_3 = filtered_columns_3.apply(lambda x: x.str.contains('DALL-E').sum())
count_deepdream_3 = filtered_columns_3.apply(lambda x: x.str.contains('DeepDream').sum())
count_dreamstudio_3 = filtered_columns_3.apply(lambda x: x.str.contains('DreamStudio').sum())

# Step 3: Plot the pie charts with specific colors
colors = ['pink', 'darkred', 'darkcyan']  # Blue for DALL-E, Pink for DeepDream, Red for DreamStudio

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Pie chart for Part A questions
axs[0].pie([count_dall_e_3.sum(), count_deepdream_3.sum(), count_dreamstudio_3.sum()], 
           labels=['DALL-E', 'DeepDream', 'DreamStudio'], 
           autopct='%1.1f%%', startangle=90, colors=colors)
axs[0].set_title('Part A Questions')

# Pie chart for Part B questions
axs[1].pie([count_dall_e_2.sum(), count_deepdream_2.sum(), count_dreamstudio_2.sum()], 
           labels=['DALL-E', 'DeepDream', 'DreamStudio'], 
           autopct='%1.1f%%', startangle=90, colors=colors)
axs[1].set_title('Part B Questions')

# Pie chart for Part C questions
axs[2].pie([count_dall_e_1.sum(), count_deepdream_1.sum(), count_dreamstudio_1.sum()], 
           labels=['DALL-E', 'DeepDream', 'DreamStudio'], 
           autopct='%1.1f%%', startangle=90, colors=colors)
axs[2].set_title('Part C Questions')

plt.show()
```
